```python
# =========================
# IMPORT LIBRARIES
# =========================
import pandas as pd
import numpy as np
import joblib
import os

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score

# =========================
# LOAD DATASET
# =========================
df = pd.read_csv("job_salary_prediction_dataset1.csv")

# =========================
# CLEAN COLUMN NAMES
# =========================
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

print("Dataset Columns:")
print(df.columns)

# =========================
# REMOVE MISSING VALUES
# =========================
df.dropna(inplace=True)

# =========================
# CHECK TARGET COLUMN
# =========================
if 'salary' not in df.columns:
    print("ERROR: salary column not found")
    exit()

# =========================
# ENCODE CATEGORICAL COLUMNS
# =========================
cat_cols = df.select_dtypes(include=['object']).columns.tolist()

# Remove target column if needed
if 'salary' in cat_cols:
    cat_cols.remove('salary')

# One-hot encoding
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)

# =========================
# FEATURES & TARGET
# =========================
X = df.drop('salary', axis=1)
y = df['salary']

# =========================
# FEATURE SCALING
# =========================
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# =========================
# TRAIN TEST SPLIT
# =========================
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# =========================
# TRAIN KNN MODEL
# =========================
model = KNeighborsRegressor(n_neighbors=5)

model.fit(X_train, y_train)

# =========================
# PREDICTION
# =========================
y_pred = model.predict(X_test)

# =========================
# EVALUATION
# =========================
score = r2_score(y_test, y_pred)

print("\nR2 Score:", score)

# =========================
# SAVE MODEL FILES
# =========================
joblib.dump(model, "knn_model.pkl",compress=9)

joblib.dump(X.columns.tolist(), "columns.pkl")

joblib.dump(scaler, "scaler.pkl")

# =========================
# VERIFY FILES
# =========================
print("\nFiles Created Successfully:")
print(os.listdir())
```
